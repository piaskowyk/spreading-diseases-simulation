```python
df = pd.read_csv("19-06-2021_21-25-08.csv")
```

```python
colors = ['#505050', '#40ff40', '#ff0d29', '#000000', '#9129a6', '#ffb60d', '#29a629']
```

```python
plt.figure(figsize=(10, 7))
plt.title("title")
plt.xlabel("x")
plt.ylabel("y")

# plt.plot(df['epoch'], df['all_agents'], '-', label='all_agents', color=colors[0])
plt.plot(df['epoch'], df['health'], '-', label='health', color=colors[1])
plt.plot(df['epoch'], df['sick'], '-', label='sick', color=colors[2])
plt.plot(df['epoch'], df['death'], '-', label='death', color=colors[3])
plt.plot(df['epoch'], df['quarantine'], '-', label='quarantine', color=colors[4])
plt.plot(df['epoch'], df['infected'], '-', label='infected', color=colors[5])
plt.plot(df['epoch'], df['recovered'], '-', label='recovered', color=colors[6])

# plt.ylim(0, 100)
plt.ylim(0)
plt.legend()
plt.tight_layout()
plt.grid()

plt.show()
```
